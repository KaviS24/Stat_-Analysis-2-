# **Label Encoding**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the Excel file into a DataFrame
file_path = "/content/Test File.xlsx"
df = pd.read_excel(file_path)

# Define categorical columns
categorical_columns = ["Country", "Category", "Manufacture"]

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Create a copy of the original DataFrame
df1 = df.copy()

# Perform label encoding for each categorical column
for col in categorical_columns:
    if col in df1.columns:
        df1[col + '_encoded'] = label_encoder.fit_transform(df1[col])

# Display the encoded DataFrame
print("Encoded DataFrame:")
print(df1)
```

```python
# View data types of each column in the DataFrame
print("Data types of each column:")
print(df.dtypes)
```

```python
# Exclude specified columns
columns_to_exclude = ["Country ", "Category ", "Manufacture"]
df_subset = df.drop(columns=columns_to_exclude)

# Convert remaining columns to integers
df_subset = df_subset.astype(int)

# View data types after conversion
print("Data types of each column after conversion:")
print(df_subset.dtypes)
```

# **Visualizing Country against all the other variables, here we obtain insights towards comparison of all the variables against each country**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of variables to plot against Country
variables = [
    'Value _BPS', 'MAT Average Price_ BPS ', 'MAT Average PriceIndex_ BPS ',
    'MAT Q4 2023 (MAT) Attractive Pack_bps ', 'MAT Q4 2023 (MAT) Better Quality Index_BPS',
    'MAT Brand Power Score _ bps ', 'Different Index _bps', 'Meaning ful Index _bps',
    'Percived Price Index _bps', 'Perceived Worth Index _BPS', 'Premium Index _ BPS ',
    'Absolute _bps ', 'Slp - Profile_BPS', 'Salient Index', 'TBCA_BPS', 'TDP_Share',
    'PD TY_BPS', 'Fair Share TY_BP', 'Penetration _BPS ', 'Relative Penetration ',
    'Repeat Penetrartion ', 'SOM%', 'SOS%', 'SOS/SOM', '% Incremental ',
    'UL Index Innovation Value Sale', 'Innovation Fair Share ', 'MAT Weight ',
    'MAT Weighted Brand Average Rating _BPS '
]

# Determine the number of rows and columns for subplots
n_cols = 3
n_rows = (len(variables) + n_cols - 1) // n_cols

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each variable against Country
for i, var in enumerate(variables):
    sns.barplot(x='Country ', y=var, data=df, ax=axes[i])
    axes[i].set_title(f'{var} by Country')
    axes[i].set_xlabel('Country')
    axes[i].set_ylabel(var)
    axes[i].tick_params(axis='x', rotation=90)

# Remove any unused subplots
for i in range(len(variables), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Find the highest Value _BPS for each category
max_value_bps_per_category = df.loc[df.groupby('Category ')['Value _BPS'].idxmax()]

# Plot the highest Value _BPS for each category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category ', y='Value _BPS', data=max_value_bps_per_category)
plt.title('Highest Value _BPS for Each Category')
plt.xlabel('Category')
plt.ylabel('Value _BPS')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 1: Identify the top 5 countries with the highest Value _BPS
top_5_countries = df.groupby('Country ')['Value _BPS'].max().nlargest(5).index

# Step 2: Filter the dataset to include only these top 5 countries
top_5_df = df[df['Country '].isin(top_5_countries)]

# Step 3: Create a bar plot to visualize Value _BPS for each category within these top 5 countries
plt.figure(figsize=(14, 8))
sns.barplot(x='Country ', y='Value _BPS', hue='Category ', data=top_5_df)
plt.title('Value _BPS for Categories in Top 5 Countries')
plt.xlabel('Country ')
plt.ylabel('Value _BPS')
plt.xticks(rotation=45)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
